Importing necessary packages

```python
import cv2
from imutils import paths
import matplotlib.pyplot as plt
import numpy as np
```

Accessing images

```python
imagePaths = list(paths.list_images('images/'))
print(imagePaths)
arrowLabels=['left','right']
print(arrowLabels)
```

Plotting our image

```python
arrowtemplate={}
fig=plt.figure(figsize=(10,10))
for i,(image,label) in enumerate(zip(imagePaths,arrowLabels)):
    image=cv2.imread(image)
    ref_image=cv2.resize(image,(100,100))
    arrowtemplate[label]=ref_image
    #print(image.shape)
    fig.add_subplot(1, 2, i+1)
    plt.imshow(ref_image)
plt.show()
    
```

```python

inputImage=cv2.imread('images/right.jpg')
inputImage=cv2.resize(inputImage,(100,100))
scores=[]
for label,image in arrowtemplate.items():
    result = cv2.matchTemplate(inputImage, image,cv2.TM_CCOEFF)
    (_, score, _, _) = cv2.minMaxLoc(result)
    scores.append(score)
```

```python
arrowindex=np.argmax(scores)
arrowindex
```
